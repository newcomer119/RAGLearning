CHROMA 

Chroma is a AI-NATIVE open-source vector database focused on developer productivity and happiness  

```python
## building a sample vector db
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import HuggingFaceEmbeddings  # runs locally, no Ollama needed
from langchain_text_splitters import RecursiveCharacterTextSplitter 
```

```python
loader = TextLoader("speech.txt")
data = loader.load()
data
```

```python
## Text Splitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
splits = text_splitter.split_documents(data)
```

```python
# HuggingFace embeddings run locally (no Ollama server required)
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectordb = Chroma.from_documents(documents=splits, embedding=embedding)
vectordb
```

```python
query = "What does the speaker believe is the main reason the United States should enter the war"
docs = vectordb.similarity_search(query)
docs[0].page_content
```
